longest length of consigutive substing


```python
s="abcdbcacdsaefghib"
left=0
result=set()
longest=0
for right in range(len(s)):
    while s[right] in result:
        result.remove(s[left])
        left+=1
    result.add(s[right])
    longest=max(longest,right-left+1)
print(longest)
```

two pointer pattern problem 
this is most important concept a programmer should understand basically it is a two pointers pointing a values in array, why we use this because using two for loop is increase time and space complexity this is best way to write program

In the given array we have to find sum is 8 arr = [1,2,3,4,5,6]

```python
arr = [1,2,3,4,5,6]
sum=0
target=8
i=0
j=len(arr)-1
while i<j:
    sum=arr[i]+arr[j]
    if sum==target:
        print(f'{arr[i],arr[j]}={sum}')
        break
    elif sum<target:
        i+=1
    else:
        j-=1         
```

now learn how to remove duplicates using two pointers 

```python
arr=[1,2,2,3,3,3,4,4,4,4,5]
result=[]
i=0
j=1
while j<len(arr):
    if arr[i]!=arr[j]:
        result.append(arr[i])
        i=j
    j+=1
print(result)    
```
